```python
driver.get("http://moodle.alphacrc.com:5081/mod/book/view.php?id=708")
```

```python
column_names = ["Book", "Chapter", "Tags", "URL"]
df = pd.DataFrame(columns=column_names)
```

```python
def scrape_data():
    book = driver.find_element(By.XPATH, "//h2")
    chapter = driver.find_element(By.XPATH, "//div[contains(@class,'box py-3')]//h3[1]")
    tags = driver.find_elements(By.XPATH, "(//a[@class='badge badge-info '])")
    the_url = driver.current_url
    tags_list = []
    for tag in tags:
        tags_list.append(tag.text)
    new_row = {"Book": book.text,
                "Chapter": chapter.text,
                "Tags": tags_list,
                "URL": the_url}
    df.loc[len(df)] = new_row

def element_found(xpath):
    try:
        driver.find_element(By.XPATH, xpath)
        return True
    except:
        return False
```

```python
scrape_data()
while element_found("(//a[@class='booknext']//i)[1]") == True:
    next = driver.find_element(By.XPATH, "(//a[@class='booknext']//i)[1]")
    next.click()
    scrape_data()

print("all done")
```
